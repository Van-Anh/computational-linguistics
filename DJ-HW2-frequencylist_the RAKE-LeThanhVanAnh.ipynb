```python
from nltk import wordpunct_tokenize, word_tokenize, pos_tag, sent_tokenize
from nltk.stem import WordNetLemmatizer
import pymorphy2 as pm
import re
```

```python
#creat a list of namefiles
doc=['Air_pollution_2.txt','Air_pollution_3.txt',
     'Air_pollution_4.txt','Air_pollution_5.txt','Air_pollution_6.txt'] #Air_pollution_1.txt',
```

```python
#read all text files
text=''
for t in doc:
    with open(t, encoding='utf-8') as f:
        text=text+(f.read())
```

```python
#function to lemmatize english text
def lemmatize_nltk(fname, text):
    """
    fname: название файла, куда будет записан результат лемматизации
    text: текст для анализа
    """
    tokens = word_tokenize(text)
    tagged = pos_tag(tokens)
    wl = WordNetLemmatizer()
    
    with open(fname, 'w', encoding='utf-8') as outfile:
        for i in range(len(tagged)):
            if tagged[i][1].startswith('N'):
                outfile.write(wl.lemmatize(tagged[i][0], pos='n') + ' ')
            elif tagged[i][1].startswith('V') or tagged[i][1].startswith('M'):
                outfile.write(wl.lemmatize(tagged[i][0], pos='v') + ' ')
            else:
                outfile.write(tagged[i][0] + ' ')
```

```python
#text - preprocessing
lemma=lemmatize_nltk('text_lemma.txt',text)
```

```python
#read the processed text
with open('text_lemma.txt', encoding='utf-8') as f:
    text_lemma = f.read()
```

```python
#read text file with english stopwords
with open('en_stopwords.txt','r',encoding='utf-8') as f:
    stopwords=f.read().split('\n')
```

```python
from collections import Counter
tokens=re.split(r'\W+', text_lemma)
# get rid of stop words. leave only those tokens that are not contained in the stopword list
tokens_new = [el for el in tokens if el.lower() not in stopwords]
#Create word frequency lists and count
Counter(tokens_new).most_common()
```

```python
from rake_nltk import Metric, Rake
r = Rake(
    stopwords=stopwords,
    ranking_metric=Metric.DEGREE_TO_FREQUENCY_RATIO,
    min_length=2,
    max_length=5
)
r.extract_keywords_from_text(text_lemma)
 # Get keyword phrases ranked highest to lowest with scores
r.get_ranked_phrases_with_scores()
```

#### <hr />The basic frequency list and the keyword list with the RAKE algorithm are completely different. 

*The words in the basic frequency list are single words.Through the basic frequency list keyword list we can know that the documents are about air pollution in general. We have to manually link the keywords in order to get common words and concepts related to air pollution.  Deep understanding of the content is not possible.<br> 

*The keyword list with the RAKE algorithm includes mostly content words phrases. The RAKE algorithm is not simply get words from documents but get information from the document. More specifically, the phrases are significance and help them more understand about content of documents.<br>
Examples: <br>
-information about health effect of air pollution: 'cervical neoplasia among hpvinfected woman', 'leukemia beryllium compound burn coal', 'widely conduct demographic health surveys', '10 people breathe unhealthy air', 'estimated 900 death per year', etc.<br>
-information about the values of particulate matter (PM): '20 microgram per cubic meter', 'pm10 particle per cubic meter', 'microgram per cubic meter', m3 annual mean 200 μg ',' guideline value o3 100 μg ', etc.
